```python
import pandas as pd
import numpy as np
data=pd.read_csv('C:/Users/louis/Documents/Msc data science/Essec/Big Data Analytics/FIFA_dataset.csv',delimiter=",")
line=data.shape[0]
data.dropna()
for i in range(line):
    if str(data.loc[i,'overall'])=="nan":
        data=data.drop(i)
Country={'Argentina':0,'Australia':1, 'Belgium':2,'Brazil':3,'Colombia':4,'Costa Rica':5,'Croatia':6,'Denmark':7,'Egypt':8,'England':9,'France':10,'Germany':11,'Iceland':12,'Iran':13,'Japan':14,'Korea Republic':15,'Mexico':16,'Morocco':17,'Nigeria':18,'Panama':19,'Peru':20,'Poland':21,'Portugal':22,'Russia':23,'Saudi Arabia':24,'Senegal':25,'Serbia':26,'Spain':27,'Sweden':28,'Switzerland':29,'Tunisia':30,'Uruguay':31}
data['Country']=data['Country'].map(Country)
 
        
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import cross_validation
from sklearn.model_selection import cross_val_score

X_pred=data.loc[data['Year']==2018]
X_pred2014=data.loc[data['Year']==2014]
X_train=data.loc[data['Year']!=2018]
X_train2014=data.loc[data['Year']<2014]
#print(X_train)
cols=[0,8,17]
#cols=[0,1,2,8,14,15,16,17]
y_train=X_train['Goal.Diiference']
y_train2014=X_train2014['Goal.Diiference']

X_train=X_train.drop(data.columns[cols],axis=1)
X_pred=X_pred.drop(data.columns[cols],axis=1)

X_train2014=X_train2014.drop(data.columns[cols],axis=1)
X_pred2014=X_pred2014.drop(data.columns[cols],axis=1)
#print(X_train2014)
```

```python
reg=LinearRegression().fit(X_train,y_train)
scores = cross_val_score(reg, X_train, y_train, cv=10)
pred=reg.predict(X_pred)
pred
reg2014=LinearRegression().fit(X_train2014,y_train2014)
scores2014 = cross_val_score(reg2014, X_train2014, y_train2014, cv=10)
pred2014=reg2014.predict(X_pred2014)
pred2014
```
